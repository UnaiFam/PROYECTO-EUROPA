```python
df2024_2=df2024.T
df2024_2
```

```python
df2024_3=df2024_2.iloc[10::3]
df2024_3.set_index(partidos2024, inplace=True)
df2024_3["suma"]=df2024_3.sum(axis=1)
df2024_3

```

```python
sumatorio=df2024_3["suma"].sum()
sumatorio
```

```python
df2024_3["% de votos total"]=df2024_3["suma"]/sumatorio*100
df2024_3
```

```python



dfpartidos = df2024_3[df2024_3['% de votos total'] > 2]


print(dfpartidos)
```

```python
df2024_otros= df2024_3[df2024_3['% de votos total'] < 2]
df2024_otros1=df2024_otros.sum(axis=0)
df2024_otros1
```

```python
df2024_T=dfpartidos.T
df2024_T["Otros"]=df2024_otros1
df2024_final=df2024_T.T
```
