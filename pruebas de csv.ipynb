```python
import pandas as pd
import numpy as np
# plotly
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)



# word cloud library
import wordcloud

# matplotlib
import matplotlib.pyplot as plt
```

```python
partidoscsv=pd.read_csv("E24 CSV\MunE24_c.csv", sep=";", encoding='latin-1')
partidos=partidoscsv.columns[10:-1:3]
partidos
```

```python
df=pd.read_csv("E24 CSV\MunE24_c.csv", sep=";", encoding='latin-1', header= 1)
df
```

```python
df.iloc[0][10:92]
```

```python
df2=df.T
df2
```

```python
df3=df2.iloc[10::3]
df3.set_index(partidos, inplace=True)
df3["suma"]=df3.sum(axis=1)
df3

```

```python
sumatorio=df3["suma"].sum()
sumatorio
```

```python
df3["% d votos total"]=df3["suma"]/sumatorio*100
df3
```

```python



dfpartidos = df3[df3['% d votos total'] > 2]


print(dfpartidos)
```

```python
df_otros= df3[df3['% d votos total'] < 2]
df_otros1=df_otros.sum(axis=0)
df_otros1
```

```python
df_T=dfpartidos.T
df_T["Otros"]=df_otros1
df_final=df_T.T
```

```python
df_final
```

```python
fig = {
  "data": [
    {
      "values": df_final["suma"],
      "labels": df_final.index,
      "domain": {"x": [0, .5]},
      "name": "Number Of Students Rates",
      "hoverinfo":"label+percent+name",
      "hole": 0.3,
      "type": "pie"
    },],
  "layout": {
        "title":"votos",
        "annotations": [
            { "font": { "size": 20},
              "showarrow": False,
              "text": "",
                "x": 0.13,
                "y": 1.1
            },
        ]
    }
}
iplot(fig)
```
