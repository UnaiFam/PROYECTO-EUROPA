```python
import pandas as pd
```

```python
ruta_archivo = "Fuentes de datos demograficos/Partidos Grupos.csv"

# Carga el archivo con los parámetros correctos
df = pd.read_csv(ruta_archivo, sep=';', encoding='latin1', decimal=',')

# Guarda como nuevo CSV (con coma como separador por defecto)
df.to_csv("datos_partido_grupo_2024", index=False)

# Imprime información
print(df.shape)
df.head(50)
```
