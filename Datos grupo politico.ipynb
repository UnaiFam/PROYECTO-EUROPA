```python
ruta_archivo_2024 = "Fuentes de datos demograficos - votos/Partidos Grupos_2024.csv"
ruta_archivo_2019 = "Fuentes de datos demograficos - votos/Partidos Grupos_2019.csv"
ruta_archivo_2014 = "Fuentes de datos demograficos - votos/Partidos Grupos_2014.csv"

df_2024 = pd.read_csv(ruta_archivo_2024, sep=';', encoding='utf-8-sig', decimal=',')
df_2019 = pd.read_csv(ruta_archivo_2019, sep=';', encoding='utf-8-sig', decimal=',')
df_2014 = pd.read_csv(ruta_archivo_2014, sep=';', encoding='utf-8-sig', decimal=',')

df_2024['Año'] = 2024
df_2019['Año'] = 2019
df_2014['Año'] = 2014

df_combined = pd.concat([df_2024, df_2019, df_2014], ignore_index=True)

df_combined.to_csv("Partidos_Politicos.csv", index=False)

print(df_combined.shape)
df_combined.head(50)
```
